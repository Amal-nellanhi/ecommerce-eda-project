```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../E-commerce_dataset.csv')
df.head()
```

```python
df = df.rename(columns={"accessed_Ffom": "Browser"})
```

```python
df.head()
```

```python
print(df["Browser"].unique())
```

```python
df["Browser"] = df["Browser"].replace({"SafFRi": "Safari"})
```

```python
browser_sales = df.groupby("Browser")["sales"].sum()

plt.figure(figsize=(8,8))  
browser_sales.plot(
    kind="pie",
    autopct='%1.1f%%',
    startangle=140,
    labels=browser_sales.index  
)

plt.title("Sales by Browser")
plt.ylabel("")  
plt.axis("equal")  
plt.show()
```
