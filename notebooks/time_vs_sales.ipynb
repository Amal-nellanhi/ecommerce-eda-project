## Introduction
The purpose of this analysis is to investigate the relationship between the time users spend on the website (duration_secs) and the amount they purchase (sales). This analysis helps us determine if higher engagement time correlates with increased revenue, which can inform strategies for improving user retention and content engagement.

## Methodology
1. **imported the dataset** using pandas and visualization libraries (Matplotlib and Seaborn).

2. **Selected relevant data** by filtering for the duration_seconds and sales columns.

3. **Cleaned the dataset** to ensure accuracy:

   - Removed rows with missing or null values in the selected columns to prevent calculation errors.

   - Ensured data types were numeric for accurate plotting.

4. **Calculated Correlation** to find the numerical strength of the relationship.

5. **Visualized** the data using:

    - A **Scatter Plot** to observe the distribution and trends between time spent and sales.

    - A **Correlation Heatmap** to graphically display the correlation coefficient.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../E-commerce_dataset.csv')
df.head()
```

```python
df_subset = df[['duration_(secs)', 'sales']]
df_subset = df_subset.dropna()
print(df_subset)
```

```python
correlation_value = df_subset['duration_(secs)'].corr(df_subset['sales'])
print(f"Correlation between Time Spent and Sales: {correlation_value:.2f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['duration_(secs)'], df['sales'])
plt.title("Time Spent vs Sales")
plt.xlabel("Duration (seconds)")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df_subset.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap: Duration vs Sales")
plt.show()
```

## Conclusion
After cleaning the dataset and visualizing the relationship between duration_(secs) and sales, I analyzed the scatter plot and correlation coefficient to understand user behavior.

From the visualization, it is clear that there is no significant linear relationship between the time a user spends on the website and the amount they purchase. The correlation coefficient is approximately **-0.0011**, which is extremely close to zero, and the scatter plot shows a random distribution of points rather than a clear upward trend. This indicates that keeping users on the website longer does not automatically lead to higher sales for this dataset.
